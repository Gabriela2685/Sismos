```python
import pandas as pd
```

```python
df_sismos = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vRnj9YeriIZMfnU2XI6Wubq9YGeripEgWYr3HB42GaON2iBu6uz98WhFuhTh4xRYA/pub?output=csv')
```

```python
df_sismos
```

```python
df_sismos['fecha_hora'] = pd.to_datetime(df_sismos['Fecha UTC'] + ' ' + df_sismos['Hora UTC'])

# Calcula los tiempos de interllegada
df_sismos['tiempo_interllegada'] = df_sismos['fecha_hora'].diff()

# Convierte los tiempos de interllegada a días
df_sismos['tiempo_interllegada_dias'] = df_sismos['tiempo_interllegada'].dt.total_seconds() / (24 * 3600)

# Acumula los tiempos de interllegada
df_sismos['tiempo_acumulado'] = df_sismos['tiempo_interllegada_dias'].cumsum()

print(df_sismos[['fecha_hora', 'tiempo_acumulado']])
```

```python
# Filtrar los sismos con magnitud mayor a 7
sismos_mayores_7 = df_sismos[df_sismos['Magnitud'] > 7]
df_ord = sismos_mayores_7.sort_values(by='tiempo_interllegada_dias', ascending=False)
df_ord
```

```python
print('Total:',sismos_mayores_7.shape[0])
```

```python
# Crear un nuevo DataFrame con solo las dos columnas deseadas
sismos_seleccionados = sismos_mayores_7[['Magnitud', 'tiempo_interllegadas_dias']]

# Mostrar el nuevo DataFrame
print(sismos_seleccionados)
```
